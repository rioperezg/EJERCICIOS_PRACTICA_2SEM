```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import chardet
from matplotlib.pyplot import rcParams

file_path = "pokemon.csv"
with open(file_path, "rb") as f:
    result = chardet.detect(f.read())

# Leer el archivo CSV utilizando la codificación detectada
df_pokemon = pd.read_csv(file_path, encoding=result["encoding"], on_bad_lines="skip")
```

```python
# Conteo de valores perdidos/faltantes  
df_pokemon.isna().sum()
```

```python
#SOLO LAS COLUMNAS QUE INTERESAN
df_pokemon_filt= df_pokemon[['name','base_total','capture_rate']]
df_pokemon_filt
```

```python
df_pokemon_filt['capture_rate'] = pd.to_numeric(df_pokemon_filt['capture_rate'], errors='coerce')

# Normaliza la columna 'capture_rate'
minimo = df_pokemon_filt['capture_rate'].min()
maximo = df_pokemon_filt['capture_rate'].max()

df_pokemon_filt['capture_rate_normalizada'] = (df_pokemon_filt['capture_rate'] - minimo) / (maximo - minimo)
df_pokemon_filt['capture_rate_normalizada'] = df_pokemon_filt['capture_rate_normalizada'] * (10 - 0) + 0

df_pokemon_filt.head()
```

```python

df_pokemon_filt = df_pokemon_filt.sort_values(by=['base_total', 'capture_rate_normalizada'], ascending=[False, True])
df_pokemon_filt
```

```python
# Asegúrate de que el índice de df_pokemon_filt esté restablecido
df_pokemon_filt = df_pokemon_filt.reset_index(drop=True)
df_pokemon_filt
```

```python

# Elimina los Pokémon con un capture_rate_normalizada mayor que 10
df_pokemon_filt = df_pokemon_filt[df_pokemon_filt['capture_rate_normalizada'] <= 10]
df_pokemon_filt
```

```python

# Ordena el DataFrame por capture_rate_normalizada
df_pokemon_filt = df_pokemon_filt.sort_values('capture_rate_normalizada')
df_pokemon_filt
```

```python

# Crea una lista de las sumas acumulativas de capture_rate_normalizada
sumas_acumulativas = df_pokemon_filt['capture_rate_normalizada'].cumsum()
print(sumas_acumulativas)
```

```python
def es_solucion(solucion, capacidad):
    total = 0
    for elemento in solucion:
        total = round(total + elemento|1| 2)
    if (total == capacidad):
            return True
    else:
            return False
def mochila(conjunto_candidatos, capacidad):
    solucion = []
    restante = capacidad 
    while conjunto_candidatos & ~ es_solucion(solucion, capacidad):
        dato = conjunto_candidatos.pop()
        if(dato[1] <= restante):
            solucion.append(dato)
            restante = round(restante . dato[1], 2)
    return solucion
# Another common operation is the use of boolean vectors to filter the data. The operators are: | for or, & for and, and ~ for not. These must 
# be grouped by using parentheses.
mochila(df_pokemon_filt, 6)
```
